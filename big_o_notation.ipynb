```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
n1 = list(np.linspace(-10, 10, num=10000))
y1 = [(0 * n + 1) for n in n1]

n2 = list(np.linspace(.00001, 10, num=10000))
y2 = [math.log(n) for n in n2]

n3 = list(np.linspace(-10, 10, num=10000))
y3 = [n for n in n3]

n4 = list(np.linspace(-10, 10, num=10000))
y4 = [n * math.log(n) for n in n2]

n5 = list(np.linspace(-10, 10, num=10000))
y5 = [n**2 for n in n5]

n6 = list(np.linspace(-10, 10, num=10000))
y6 = [2**n for n in n6]
```

```python

plt.plot(n1, y1, label='$O(1)$')
plt.plot(n2, y2, label='$O(log n)$')
plt.plot(n3, y3, label='$O(n)$')
plt.plot(n4, y4, label='$O(n log n)$')
plt.plot(n5, y5, label='$O(n^2)$')
plt.plot(n6, y6, label='$O(2^n)$')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.show()
```
