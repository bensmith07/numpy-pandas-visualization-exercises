# Matplotlib Exercises

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

#### 1. Use matplotlib to plot the following equation: 

$y = x^2 - x + 2$

```python
x = list(range(-100, 101))
y = [(x**2 - x + 2) for x in x]
```

```python
plt.plot(x, y)
plt.ylim(-1000, 10000)
plt.text(-9, -500, '(0, 0)')
plt.show()
```

#### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- $y = \sqrt{x}$
- $y = x^3$
- $y = 2^x$
- $y = 1/(x + 1)$

```python
x1 = list(np.linspace(0, 10, num=10000))
x2 = list(np.linspace(-10, 10, num=10000))
x3 = list(np.linspace(-10, 10, num=10000))

x4_a = list(np.linspace(-1, 10, num=10000))
x4_b = list(np.linspace(-10, -1, num=10000))
x4_a.remove(-1)
x4_b.remove(-1)

x_min = -10
x_max = 10
y_min = -10
y_max = 10
```

```python
# y = sqrt(x)
plt.plot(x1, [math.sqrt(x1) for x1 in x1])
plt.title('$y = \sqrt{x}$')
plt.show()
```

```python
# y = x**3
plt.plot(x2, [x2**3 for x2 in x2])
plt.title('$y = x^3$')
plt.show()
```

```python
# y = 2**x
plt.plot(x3, [2**x3 for x3 in x3])
plt.title('$y = 2^x$')
plt.show()
```

```python
# y = 1/(x+1)
plt.plot(x4_a, [1/(x4_a+1) for x4_a in x4_a], c='blue')
plt.plot(x4_b, [1/(x4_b+1) for x4_b in x4_b], c='blue')
plt.title('$y = 1/(x + 1)$')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

#### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
n_rows = 2
n_cols = 2

# 1st subplot (y = sqrt(x))
x = list(range(1, 20))
plt.subplot(n_rows, n_cols, 1)
plt.plot(x1, [math.sqrt(x1) for x1 in x1])
plt.title('$y = \sqrt{x}$')

# 2nd subplot (y = x**3)
x = list(range(-20, 20))
plt.subplot(n_rows, n_cols, 2)
plt.plot(x2, [x2**3 for x2 in x2])
plt.title('$y = x^3$')

# 3rd subplot (y = 2**x)
x = list(range(-20, 20))
plt.subplot(n_rows, n_cols, 3)
plt.plot(x3, [2**x3 for x3 in x3])
plt.title('$y = 2^x$')

# 4th subplot (y = 1/(x+1))
x = list(range(-20, -1)) + list(range(20))
plt.subplot(n_rows, n_cols, 4)
plt.plot(x4_a, [1/(x4_a+1) for x4_a in x4_a], c='blue')
plt.plot(x4_b, [1/(x4_b+1) for x4_b in x4_b], c='blue')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.title('$y = 1/(x + 1)$')

plt.subplots_adjust(hspace=.5, wspace=.4)

plt.show()
```

#### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure()

#1 (y = sqrt(x))
x = list(range(1, 20))
plt.plot(x1, [math.sqrt(x1) for x1 in x1], label='$y = \sqrt{x}$')

#2 (y = x**3)
x = list(range(-20, 20))
plt.plot(x2, [x2**3 for x2 in x2], label='$y = x^3$')

#3 (y = 2**x)
x = list(range(-20, 20))
plt.plot(x3, [2**x3 for x3 in x3], label='$y = 2^x$')

#4 (y = 1/(x+1))
x = list(range(-20, -1)) + list(range(20))
plt.plot(x4_a, [1/(x4_a+1) for x4_a in x4_a], c='red', label='$y = 1/(x + 1)$')
plt.plot(x4_b, [1/(x4_b+1) for x4_b in x4_b], c='red')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title('FUNctions', fontsize=20)
plt.legend(loc='lower right')

plt.show()

# - $y = \sqrt{x}$
# - $y = x^3$
# - $y = 2^x$
# - $y = 1/(x + 1)$
```
