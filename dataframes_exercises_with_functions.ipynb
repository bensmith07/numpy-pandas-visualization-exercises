#### Copy the code from the lesson to create a dataframe full of student grades.

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

grades_data = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
grades_data.head()
```

### 1a.  Original Exercise

#### Create a column named passing_english that indicates whether each student has a passing grade in english.

```python
# create the new column
grades_data['passing_english'] = grades_data.english > 70

# confirm that it worked
grades_data.head()
```

### 1a. Using a Function

#### Define a function that accepts the dataframe of grades, a subject, and a minimum grade as arguments. The function should return a new version of the dataframe containing a column of boolean values that indicates whether the student's grade in that subject is above the minimum value

```python
# this one is kind of complicated. 
# hopefully won't be replicated on the exam

def create_passing_column(df, subject, min_grade):
    new_column_name = f'passing_{subject}'
    df[new_column_name] = df[subject] >= min_grade
    return df
```

#### Use the function defined above to create a new column indicating whether each student has a passing grade in math. 

```python
# create the function by calling the function
# and reassigning the result to our original dataframe variable
grades_data = create_passing_column(grades_data, 'math', 70)
```

```python
grades_data.head()
```

### 1e. Original Exercise

#### Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
grades_data['overall_grade'] = grades_data[['math', 'english', 'reading']].mean(axis=1).round(2)
```

```python
grades_data.head()
```

### 1e. Using a function

#### Define a function that takes in a dataframe and a list of columns, and a new column name, then returns a new dataframe with a new column that is the average of the given columns. 

```python
def create_average_column(data, new_column_name, column_list):
    data[new_column_name] = data[column_list].mean(axis=1).round(2)
    return data
```

#### Use the function defined above to add a column called overall_grade to the grades data that contains that average of the math, english, and reading grades. 

```python
# first, drop the overall_grade column that we created before
# so that we can create it again
grades_data = grades_data.drop(columns='overall_grade')
grades_data.head()
```

```python
grades_data = create_average_column(grades_data, 'overall_grade', ['math', 'english', 'reading'])
```

### 2a. With Function
#### Define two functions, one that returns the number of rows from a dataframe, and one that returns the number of columns from a dataframe

```python
def get_rows(df):
    return df.shape[0]

def get_cols(df):
    return df.shape[1]
```

#### Use the function defined above to output the number of rows and columns in the mpg dataframe

```python
print(f'Rows = {get_rows(mpg)}')
print(f'Columns = {get_cols(mpg)}')
```
